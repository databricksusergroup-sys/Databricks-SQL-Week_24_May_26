# Answer 1

## What is Data?

Data is a collection of raw facts, values, or information that can be processed to generate meaningful insights.

### Types of Data

- **Structured Data**
  - Organized in rows and columns
  - Example: SQL tables, Excel sheets

- **Semi-Structured Data**
  - Does not follow a strict tabular structure
  - Example: JSON, XML

- **Unstructured Data**
  - No predefined structure
  - Example: Images, videos, audio files

# Answer 2

## What is a Database?

A database is an organized collection of data stored electronically for easy access, management, and retrieval.

### Importance of Databases

- Store large amounts of data
- Enable fast data retrieval
- Reduce redundancy
- Improve data security
- Support applications and analytics

# Answer 3

## What is DBMS?

DBMS (Database Management System) is software used to create, manage, and maintain databases.

### Main Responsibilities

- Data storage
- Data retrieval
- Security management
- Backup and recovery
- Data consistency

### Examples

- MySQL
- PostgreSQL
- Oracle
- SQL Server

# Answer 5

## What is a Data Warehouse?

A Data Warehouse is a centralized system used for storing and analyzing historical business data.

### Features

- Stores structured data
- Optimized for analytics
- Supports reporting and dashboards

### Use Cases

- Business Intelligence
- Sales reporting
- Trend analysis

# Answer 6

## What is a Data Lake?

A Data Lake is a storage system that stores structured, semi-structured, and unstructured data in raw format.

### Advantages

- Handles huge volumes of data
- Flexible storage
- Useful for big data and machine learning

# Answer 7

## Data Warehouse vs Data Lake

| Feature | Data Warehouse | Data Lake |
|---|---|---|
| Data Type | Structured | All types |
| Storage | Processed data | Raw data |
| Purpose | Analytics | Big data storage |
| Users | Business analysts | Data engineers/scientists |

# Answer 8

## OLTP vs OLAP

### OLTP
- Used for daily transactions
- Fast inserts and updates
- Example: Banking systems

### OLAP
- Used for analytics and reporting
- Complex queries on large datasets
- Example: Business dashboards

# Answer 10

## Databricks Namespace Structure

```text
catalog.schema.table
```

### Catalog
Top-level container for organizing data assets.

### Schema
Logical grouping of tables inside a catalog.

### Table
Stores actual data in rows and columns.

# Answer 11

## What is SQL?

SQL (Structured Query Language) is used to communicate with databases.

### Categories of SQL Commands

- **DDL** → CREATE, ALTER, DROP
- **DML** → INSERT, UPDATE, DELETE
- **DQL** → SELECT
- **DCL** → GRANT, REVOKE
- **TCL** → COMMIT, ROLLBACK

# Answer 12

## Types of Keys

### Primary Key
Uniquely identifies each row in a table.

### Foreign Key
Creates relationships between tables.

### Candidate Key
A column eligible to become a primary key.

### Composite Key
Combination of multiple columns forming uniqueness.

### Unique Key
Ensures all values are unique.
