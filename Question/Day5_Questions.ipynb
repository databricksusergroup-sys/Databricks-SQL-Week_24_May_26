# Subquery vs JOIN

| Approach   | Best Use Case       |
| ---------- | ------------------- |
| JOIN       | Combining tables    |
| Subquery   | Stepwise filtering  |
| EXISTS     | Presence checks     |
| NOT EXISTS | Missing data checks |
