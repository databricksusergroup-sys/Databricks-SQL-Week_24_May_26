❌ Wrong:

```sql

FROM orders
GROUP BY user_id;
```

Because `total_amount` is neither grouped nor aggregated.

# 4. HAVING Clause

HAVING filters grouped results.

**Syntax:**
```sql
SELECT column_name, aggregate_function(column)
FROM table_name
GROUP BY column_name
HAVING condition;
```

### WHERE vs HAVING

| Clause | Runs Before Grouping? | Filters |
| ------ | --------------------- | ------- |
| WHERE  | Yes                   | Rows    |
| HAVING | No                    | Groups  |
